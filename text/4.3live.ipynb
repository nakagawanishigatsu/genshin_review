```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 读取CSV文件
df = pd.read_csv('4.3.csv')

# 筛选type为0的评论
comments = df[df['type'] == 0]

# 将'时间'列的字符串转换为Pandas的datetime对象
comments['datetime'] = pd.to_datetime(comments['时间'])

# 提取小时和分钟，组合成HH:MM格式的字符串
comments['time_label'] = comments['datetime'].dt.strftime('%H:%M')

# 计算每个时间点的评论数量
comment_counts_per_minute = comments.groupby('time_label').size()

# 准备绘制折线图
plt.figure(figsize=(12, 6))  # 设置图形大小
comment_counts_per_minute.index = pd.to_datetime(comment_counts_per_minute.index, format='%H:%M')

# 使用plot函数绘制折线图
plt.plot(comment_counts_per_minute.index, comment_counts_per_minute.values, marker='o')

# 设置时间格式化为小时和分钟（不显示日期）
time_format = mdates.DateFormatter('%H:%M')
plt.gca().xaxis.set_major_formatter(time_format)

# 设置图形标题和坐标轴标签
plt.title('Number of Comments per Minute')
plt.xlabel('Time of Day')
plt.ylabel('Number of Comments')

# 确保x轴的时间标签是按照时间顺序排列的
plt.xticks(rotation=45)
plt.tight_layout()  # 自动调整子图参数, 使之填充整个图像区域

# 显示图形
plt.show()
```

```python
import pandas as pd
import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt
#from imageio import imread

# 读取CSV数据
df = pd.read_csv('4.3.csv')

# 筛选type为0的行
filtered_df = df[df['type'] == 0]

# 读取哈工大停用词表
with open('hit_stopwords.txt', 'r', encoding='utf-8') as f:
    stop_words = set([line.strip() for line in f.readlines()])

# 合并所有content列的文本
all_content = ' '.join(filtered_df['message']).replace('\n', ' ').replace('\r', ' ').strip()
# 使用jieba进行分词并去除停用词
words = jieba.lcut(all_content)
words = [word for word in words if word not in stop_words and len(word.strip()) > 0]

word_freq = pd.Series(words).value_counts()

# 生成词云
wordcloud = WordCloud(width=800, height=400, font_path='simhei.ttf', background_color='white').generate_from_frequencies(word_freq)

# 显示词云
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 关闭坐标轴
plt.show()
```

```python
import pandas as pd

# 读取CSV文件
csv_file = '4.3.csv'
df = pd.read_csv(csv_file)

# 筛选type为0的行
df_filtered = df[df['type'] == 0]

# 计算每个uid的出现次数，并找出出现次数最多的15个uid
uid_counts = df_filtered['uName'].value_counts()
top_uids = uid_counts.nlargest(15).index

# 根据这些uid筛选出对应的行
top_uids_df = df_filtered[df_filtered['uName'].isin(top_uids)]

# 按uid和content分组，并计算每个组合的出现次数
grouped = top_uids_df.groupby(['uId', 'message']).size().reset_index(name='count')

# 准备输出结果
output = []

# 遍历分组结果，按照要求的格式准备输出
for uid, group in grouped.groupby('uId'):
    uid_output = []
    for index, row in group.iterrows():
        content = row['message']
        count = row['count']
        uid_output.append(f"‘{content}’*{count}")
    df1=top_uids_df[top_uids_df['uId']==uid]
    uname=df1.iloc[0]['uName']
    output.append((uid, uname,"，".join(uid_output)))

# 输出结果
for uid, uname, content_list in output:
    print(f"{int(uid)}，{uname}")
    for content in content_list.split("，"):
        print(content)
```

```python
import pandas as pd

# 读取CSV文件
csv_file = '4.3.csv'
df = pd.read_csv(csv_file)

# 筛选type为0的行
df_filtered = df[df['type'] == 0]

# 计算每个uid的出现次数，并找出出现次数最多的15个uid
uid_counts = df_filtered['uName'].value_counts()
top_uids = uid_counts.nlargest(15).index

# 根据这些uid筛选出对应的行
top_uids_df = df_filtered[df_filtered['uName'].isin(top_uids)]

# 按uid和content分组，并计算每个组合的出现次数
grouped = top_uids_df.groupby(['uId', 'message']).size().reset_index(name='count')

# 准备输出结果
output = []

# 遍历分组结果，按照要求的格式准备输出
for uid, group in grouped.groupby('uId'):
    # 计算当前uid的总条数
    total_count = len(top_uids_df[top_uids_df['uId'] == uid])
    
    uid_output = []
    for index, row in group.iterrows():
        content = row['message']
        count = row['count']
        uid_output.append(f"‘{content}’*{count}")
    
    # 获取当前uid对应的uname
    uname = top_uids_df[top_uids_df['uId'] == uid]['uName'].iloc[0]
    
    # 将结果添加到输出列表
    output.append((uid, uname, total_count, "，".join(uid_output)))

# 输出结果
for uid, uname, total_count, content_list in output:
    print(f"UID: {int(uid)}, UName: {uname}, Total Messages: {total_count}")
    print(content_list)
    print()  # 打印空行以便区分不同的uid输出
```
