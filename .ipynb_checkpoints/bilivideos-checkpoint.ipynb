```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
df = pd.read_csv('bilivideo.csv')

# 筛选出作品标题中包含“杂谈”的行
talk_videos = df[df['作品标题'].str.contains('杂谈', case=False, na=False)]

# 计算点赞率（点赞数/播放量）
talk_videos['点赞率'] = talk_videos['获赞数'] / talk_videos['播放数']

# 提取横轴标签中括号内的字符串
talk_videos['横轴标签'] = talk_videos['作品标题'].str.extract(r'《原神》拾枝杂谈-「(.+?)：.+」')

fig, ax1 = plt.subplots(figsize=(144, 17))

# 绘制点赞率折线图
color = 'tab:red'
ax1.set_xlabel('人物')
ax1.set_ylabel('点赞率', color=color)
median_like_rate = talk_videos['点赞率'].mean()  # 计算点赞率的中位数
ax1.plot(talk_videos['横轴标签'], talk_videos['点赞率'], color=color)
ax1.axhline(y=median_like_rate, color=color, linestyle='--', label='点赞率平均数')
ax1.tick_params(axis='y', labelcolor=color)

# 创建第二个轴对象
ax2 = ax1.twinx()

# 绘制弹幕数折线图
color = 'tab:blue'
ax2.set_ylabel('弹幕数', color=color)
median_danma = talk_videos['弹幕数'].mean()  # 计算弹幕数的中位数
ax2.plot(talk_videos['横轴标签'], talk_videos['弹幕数'], color=color)
ax2.axhline(y=median_danma, color=color, linestyle='--', label='弹幕数平均数')
ax2.tick_params(axis='y', labelcolor=color)

# 定制图例
lines = [plt.Line2D([0], [0], color=color, linestyle='-', linewidth=2) for color in ['tab:red', 'tab:blue']]
median_lines = [plt.Line2D([0], [0], color=color, linestyle='--', linewidth=2) for color in ['tab:red', 'tab:blue']]
labels = ['点赞率', '弹幕数']
median_labels = ['点赞率平均数', '弹幕数平均数']
ax1.legend(lines + median_lines, labels + median_labels, loc='upper left')

# 旋转横轴标签
plt.gcf().autofmt_xdate()
for label in plt.gca().get_xticklabels(which='major'):
    label.set_rotation(90)
    label.set_fontsize(26)

# 显示标题
plt.title('角色杂谈的点赞率与弹幕数')

# 显示图表
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
df = pd.read_csv('bilivideo.csv')

# 筛选出作品标题中包含“杂谈”的行
talk_videos = df[df['作品标题'].str.contains('版本PV', case=False, na=False)]

# 计算点赞率（点赞数/播放量）
talk_videos['点赞率'] = talk_videos['获赞数'] / talk_videos['播放数']

# 提取横轴标签中括号内的字符串
talk_videos['横轴标签'] = talk_videos['作品标题'].str.extract(r'(?<=《原神》)(\d\.\d)(?=版本PV)')
#talk_videos=talk_videos.dropna(axis=0,how='any')
```

```python
fig, ax1 = plt.subplots(figsize=(144, 17))

# 绘制点赞率折线图
color = 'tab:red'
ax1.set_xlabel('版本号')
ax1.set_ylabel('点赞率', color=color)
median_like_rate = talk_videos['点赞率'].mean()  # 计算点赞率的中位数
ax1.plot(talk_videos['横轴标签'], talk_videos['点赞率'], color=color)
ax1.axhline(y=median_like_rate, color=color, linestyle='--', label='点赞率平均数')
ax1.tick_params(axis='y', labelcolor=color)

# 创建第二个轴对象
ax2 = ax1.twinx()

# 绘制弹幕数折线图
color = 'tab:blue'
ax2.set_ylabel('弹幕数', color=color)
median_danma = talk_videos['弹幕数'].mean()  # 计算弹幕数的中位数
ax2.plot(talk_videos['横轴标签'], talk_videos['弹幕数'], color=color)
ax2.axhline(y=median_danma, color=color, linestyle='--', label='弹幕数平均数')
ax2.tick_params(axis='y', labelcolor=color)

# 定制图例
lines = [plt.Line2D([0], [0], color=color, linestyle='-', linewidth=2) for color in ['tab:red', 'tab:blue']]
median_lines = [plt.Line2D([0], [0], color=color, linestyle='--', linewidth=2) for color in ['tab:red', 'tab:blue']]
labels = ['点赞率', '弹幕数']
median_labels = ['点赞率平均数', '弹幕数平均数']
ax1.legend(lines + median_lines, labels + median_labels, loc='upper right')

# 旋转横轴标签
plt.gcf().autofmt_xdate()
for label in plt.gca().get_xticklabels(which='major'):
    label.set_rotation(45)
    label.set_fontsize(26)

# 显示标题
plt.title('版本PV的点赞率与弹幕数')

# 显示图表
plt.show()
```
